```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

# Crank-slider mechanism


$$x=R(1-\cos\theta)+\dfrac{R^2}{2L}\sin^2\theta$$
$$\nu=\dfrac{dx}{dt}=R\omega\left[\sin\theta+\dfrac{R}{2L}\sin2\theta\right]$$
$$a=\dfrac{d^2x}{dt^2}=R\omega^2\left[\cos\theta+\dfrac{R}{L}\cos2\theta\right]$$

# Scotch Yoke mechanism



$$x=R(1-\cos\theta)$$
$$\nu=\dfrac{dx}{dt}=R\omega\sin\theta$$
$$a=\dfrac{d^2x}{dt^2}=R\omega^2\cos\theta$$

## Input data
R = 1 ft    L = 2 ft<br/>
$\omega$ = 15 rad/s <br/>
$\theta$ = 0$^\circ$ to 360$^\circ$ with 15$^\circ$ increments

```python
r = 1
l = 2
w = 15
deg = np.linspace(0, 2*np.pi,360)
```

```python
csx = r*(1-np.cos(deg))+((r**2)/(2*l))*np.sin(deg)**2
csv = r*w*(np.sin(deg)+(r/(2*l))*np.sin(2*deg))
csa = r*w**2*(np.cos(deg)+(r/l)*np.cos(2*deg))

syx=r*(1-np.cos(deg))
syv=r*w*np.sin(deg)
sya=r*w**2*np.cos(deg)
```

### Plot Functions

```python
plt.figure(figsize=(25,25))
sns.set()
plt.subplot(431)
plt.plot(csx,label='Slider-Crank')
plt.plot(syx,label='Scotch Yoke')
plt.legend()
plt.title('Displacement vs Degree')
plt.grid(True)
plt.xlabel('Degrees')
plt.ylabel('x (ft)')

plt.subplot(432)
plt.plot(csv, label='Slider-Crank')
plt.plot(syv, label='Scotch Yoke')
plt.legend()
plt.title('Velocity vs Degree')
plt.grid(True)
plt.xlabel('Degrees')
plt.ylabel('v (ft/s)')

plt.subplot(433)
plt.plot(csa, label='Slider-Crank')
plt.plot(sya, label='Scotch Yoke')
plt.legend()
plt.title('Acceleration vs Degree')
plt.xlabel('Degrees')
plt.grid(True)
plt.ylabel('a (ft/s^2)')

plt.show()
```
